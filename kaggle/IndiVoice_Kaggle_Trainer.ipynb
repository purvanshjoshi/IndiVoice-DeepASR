## 2. Acoustic Fingerprint Visualization
Before training, let's look at what the AI 'hears'. This generates a Mel-Spectrogram (the input to Whisper) for a sample from your dataset.

```python
from IPython.display import Image, display
from src.visualize import plot_mel_spectrogram
import json
import os

# Load a sample path from the manifest
manifest_path = "data/processed/svarah_manifest.json"
if os.path.exists(manifest_path):
    with open(manifest_path, "r") as f:
        sample = json.loads(f.readline())
        audio_path = sample["audio"]
        
    print(f"Viewing sample: {audio_path}")
    plot_mel_spectrogram(audio_path, output_path="sample_spectrogram.png")
    display(Image("sample_spectrogram.png"))
else:
    print("❌ Manifest not found. Link your dataset first!")
```
